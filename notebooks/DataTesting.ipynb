```python
## 2a) points per driver in each season joined
CREATE OR REPLACE TABLE POINTS_PER_DRIVER_SEASON AS
SELECT 
    RACES.YEAR,
    DRIVERS.DRIVERID,
    DRIVERS.FORENAME,
    DRIVERS.SURNAME,
    SUM(RESULTS.POINTS) AS TOTAL_POINTS
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RESULTS AS RESULTS
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RACES AS RACES
ON 
    RESULTS.RACEID = RACES.RACEID
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.DRIVERS AS DRIVERS
ON 
    RESULTS.DRIVERID = DRIVERS.DRIVERID
GROUP BY 
    RACES.YEAR, DRIVERS.DRIVERID, DRIVERS.FORENAME, DRIVERS.SURNAME;
```

```python
## 2b) wins per constructor
CREATE OR REPLACE TABLE WINS_PER_CONSTRUCTOR AS
SELECT 
    CONSTRUCTOR.CONSTRUCTORID,
    CONSTRUCTOR.NAME AS CONSTRUCTOR_NAME,
    COUNT(*) AS TOTAL_WINS
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RESULTS AS RESULTS
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.CONSTRUCTOR AS CONSTRUCTOR
ON 
    RESULTS.CONSTRUCTORID = CONSTRUCTOR.CONSTRUCTORID
WHERE 
    RESULTS.POSITIONTEXT = '1'
GROUP BY 
    CONSTRUCTOR.CONSTRUCTORID, CONSTRUCTOR.NAME;
```

```python
## 2c fastest laps per season
CREATE OR REPLACE TABLE FASTEST_LAPS_PER_SEASON AS
SELECT 
    RACES.YEAR,
    DRIVERS.DRIVERID,
    DRIVERS.FORENAME,
    DRIVERS.SURNAME,
    COUNT(*) AS TOTAL_FASTEST_LAPS
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RESULTS AS RESULTS
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RACES AS RACES
ON 
    RESULTS.RACEID = RACES.RACEID
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.DRIVERS AS DRIVERS
ON 
    RESULTS.DRIVERID = DRIVERS.DRIVERID
WHERE 
    RESULTS.FASTESTLAP IS NOT NULL
GROUP BY 
    RACES.YEAR, DRIVERS.DRIVERID, DRIVERS.FORENAME, DRIVERS.SURNAME;
```

```python
## #total points per constructor per season
CREATE OR REPLACE TABLE TOTAL_POINTS_PER_CONSTRUCTOR_SEASON AS
SELECT 
    RACES.YEAR,
    CONSTRUCTOR.CONSTRUCTORID,
    CONSTRUCTOR.NAME AS CONSTRUCTOR_NAME,
    SUM(RESULTS.POINTS) AS TOTAL_POINTS
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RESULTS AS RESULTS
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RACES AS RACES
ON 
    RESULTS.RACEID = RACES.RACEID
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.CONSTRUCTOR AS CONSTRUCTOR
ON 
    RESULTS.CONSTRUCTORID = CONSTRUCTOR.CONSTRUCTORID
GROUP BY 
    RACES.YEAR, CONSTRUCTOR.CONSTRUCTORID, CONSTRUCTOR.NAME;
```

```python
## podium finishes
CREATE OR REPLACE TABLE PODIUM_FINISHES_PER_DRIVER AS
SELECT 
    DRIVERS.DRIVERID,
    DRIVERS.FORENAME,
    DRIVERS.SURNAME,
    COUNT(*) AS PODIUM_COUNT
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RESULTS AS RESULTS
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.DRIVERS AS DRIVERS
ON 
    RESULTS.DRIVERID = DRIVERS.DRIVERID
WHERE 
    RESULTS.POSITIONTEXT IN ('1', '2', '3')
GROUP BY 
    DRIVERS.DRIVERID, DRIVERS.FORENAME, DRIVERS.SURNAME;
```

```python
## average points per race in each season

CREATE OR REPLACE TABLE AVG_POINTS_PER_RACE_DRIVER AS
SELECT 
    DRIVERS.DRIVERID,
    DRIVERS.FORENAME,
    DRIVERS.SURNAME,
    AVG(RESULTS.POINTS) AS AVG_POINTS
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RESULTS AS RESULTS
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.DRIVERS AS DRIVERS
ON 
    RESULTS.DRIVERID = DRIVERS.DRIVERID
GROUP BY 
    DRIVERS.DRIVERID, DRIVERS.FORENAME, DRIVERS.SURNAME;
```

```python
## most frequent race winners
CREATE OR REPLACE TABLE MOST_FREQUENT_RACE_WINNERS AS
SELECT 
    DRIVERS.DRIVERID,
    DRIVERS.FORENAME,
    DRIVERS.SURNAME,
    COUNT(*) AS TOTAL_WINS
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RESULTS AS RESULTS
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.DRIVERS AS DRIVERS
ON 
    RESULTS.DRIVERID = DRIVERS.DRIVERID
WHERE 
    RESULTS.POSITIONTEXT = '1'
GROUP BY 
    DRIVERS.DRIVERID, DRIVERS.FORENAME, DRIVERS.SURNAME
ORDER BY 
    TOTAL_WINS DESC;
```

```python
## total laps per season

CREATE OR REPLACE TABLE TOTAL_LAPS_PER_SEASON AS
SELECT 
    RACES.YEAR,
    SUM(RESULTS.LAPS) AS TOTAL_LAPS
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RESULTS AS RESULTS
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RACES AS RACES
ON 
    RESULTS.RACEID = RACES.RACEID
GROUP BY 
    RACES.YEAR;
```

```python
## constructor wins per season
CREATE OR REPLACE TABLE CONSTRUCTOR_WINS_PER_SEASON AS
SELECT 
    RACES.YEAR,
    CONSTRUCTOR.CONSTRUCTORID,
    CONSTRUCTOR.NAME AS CONSTRUCTOR_NAME,
    COUNT(*) AS WINS
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RESULTS AS RESULTS
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RACES AS RACES
ON 
    RESULTS.RACEID = RACES.RACEID
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.CONSTRUCTOR AS CONSTRUCTOR
ON 
    RESULTS.CONSTRUCTORID = CONSTRUCTOR.CONSTRUCTORID
WHERE 
    RESULTS.POSITIONTEXT = '1'
GROUP BY 
    RACES.YEAR, CONSTRUCTOR.CONSTRUCTORID, CONSTRUCTOR.NAME;
```

```python
## drivers with fastest laps
CREATE OR REPLACE TABLE DRIVERS_WITH_FASTEST_LAPS AS
SELECT 
    DRIVERS.DRIVERID,
    DRIVERS.FORENAME,
    DRIVERS.SURNAME,
    COUNT(*) AS FASTEST_LAPS_COUNT
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RESULTS AS RESULTS
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.DRIVERS AS DRIVERS
ON 
    RESULTS.DRIVERID = DRIVERS.DRIVERID
WHERE 
    RESULTS.FASTESTLAP IS NOT NULL
GROUP BY 
    DRIVERS.DRIVERID, DRIVERS.FORENAME, DRIVERS.SURNAME;
```

```python
## pole positions per driver
CREATE OR REPLACE TABLE POLE_POSITIONS_PER_DRIVER AS
SELECT 
    DRIVERS.DRIVERID,
    DRIVERS.FORENAME,
    DRIVERS.SURNAME,
    COUNT(*) AS POLE_POSITIONS
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.QUALIFYING AS QUALIFYING
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.DRIVERS AS DRIVERS
ON 
    QUALIFYING.DRIVERID = DRIVERS.DRIVERID
WHERE 
    QUALIFYING.POSITION = 1
GROUP BY 
    DRIVERS.DRIVERID, DRIVERS.FORENAME, DRIVERS.SURNAME;
```

```python
## seasons highlights special

CREATE OR REPLACE TABLE SEASON_HIGHLIGHTS AS
SELECT 
    RACES.YEAR,
    COUNT(DISTINCT RACES.RACEID) AS TOTAL_RACES,
    SUM(RESULTS.POINTS) AS TOTAL_POINTS,
    COUNT(CASE WHEN RESULTS.POSITIONTEXT = '1' THEN 1 END) AS TOTAL_WINS,
    COUNT(RESULTS.FASTESTLAP) AS TOTAL_FASTEST_LAPS
FROM 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RESULTS AS RESULTS
INNER JOIN 
    ADO_ASG2_GROUP_CAT_DB.ADO_ASG2_GROUP_CAT_SCM.RACES AS RACES
ON 
    RESULTS.RACEID = RACES.RACEID
GROUP BY 
    RACES.YEAR;
```
